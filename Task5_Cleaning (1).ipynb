```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0,12,18,35,60,100],
    labels=['Child','Teen','Young Adult','Adult','Senior']
)
```

- Filled missing numeric values using median
- Filled missing categorical values using mode
